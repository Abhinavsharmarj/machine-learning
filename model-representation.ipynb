## Implement the model for linear regression with one variable

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

### input data

```python
x_train = np.array([1.0,2.0])
y_train = np.array([300.0,500.0])

print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
plt.scatter(x_train,y_train, marker = 'x', c = 'r')

plt.title('Housing Prices')
plt.ylabel('Price in 1000s of dollars')
plt.xlabel('Size in 1000 sqft')

plt.show()
```

### Use the linear regression formula:
f(x) = wx + b

```python
#Taking w and b as 100
w = 100
b = 100
```

```python
predicted_price = np.dot(x_train,w) + b
```

```python
plt.scatter(x_train,predicted_price, c = 'b', label = 'Our Prediction')
plt.scatter(x_train, y_train, marker = 'x', c = 'r', label = 'Actual Values')

plt.title('Housing Prices')
plt.ylabel('Price in 1000s of dollars')
plt.xlabel('Size in 1000 sqft')

plt.legend()
plt.show()
```

### Changing the values of w and b to see if we can fit the train data
lets try w as 200 and b as 100

```python
#Taking w as 200 and b as 100
w = 200
b = 100

#getting updated predicted prices
predicted_price = np.dot(x_train,w) + b

#plot both train data and predicted data
plt.scatter(x_train,predicted_price, c = 'b', label = 'Our Prediction')
plt.scatter(x_train, y_train, marker = 'x', c = 'r', label = 'Actual Values')

plt.title('Housing Prices')
plt.ylabel('Price in 1000s of dollars')
plt.xlabel('Size in 1000 sqft')

plt.legend()
plt.show()
```

```python
to_save = { 'x_train' : x_train,
            'y_train' : y_train,
            'w': w,
            'b': b}
```
