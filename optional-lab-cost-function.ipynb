## Cost function for linear regression with one variable

```python
import numpy as np
%matplotlib widget
import matplotlib.pyplot as plt
```

### Loading data result from previous file --> model-representation.ipynb

```python
import joblib

data = joblib.load('saved_data.joblib')
```

```python
data
```

### Problem Statement
A model to predict housing prices given the size

```python
x_train = data['x_train']
y_train = data['y_train']
```

```python
w = data['w']
b = data['b']
```

### Calculation of Cost function
Cost function is the measure of how well predicition works when compared to actual data

We are using MSE (Mean squared error) as cost function

cost function(J) for set of weights(w) and bias (b) is :

sum of { predicted y - actual y } **2 / number of examples 

where predicted y is w * x + b

```python
predicted_data = np.dot(x_train,w) + b
```

```python
total_error = 0
for i in range(len(x_train)):
    error_i = predicted_data[i] - y_train[i]
    error_i_sq = error_i**2
    total_error += error_i_sq

cost_value = total_error/len(x_train)
```

```python
cost_value
```

### We can try experimenting with changing values of w

```python
w = np.array(range(0,310, 10))
print(f"w: {w}")
```

```python
cost_function_results = []
w_i_value = []
for w_i in w:
    predicted_data = np.dot(x_train,w_i) + b

    total_error = 0
    for i in range(len(x_train)):
        error_i = predicted_data[i] - y_train[i]
        error_i_sq = error_i**2
        total_error += error_i_sq

    cost_value = total_error/len(x_train)

    cost_function_results.append(cost_value)
    w_i_value.append(w_i)
```

```python
plt.plot(w_i_value,cost_function_results)

plt.xlabel('Weights values')
plt.ylabel('Cost function value')

plt.title('Cost function vs Weights')

plt.show()
```

### Larger DataSet

```python
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y_train = np.array([250, 300, 480,  430,   630, 730,])
```

```python
#let w = 10 and b = 0 and then changing each to find the minimum
```

```python
b = 0
w = np.array(range(10,500,10))

cost_total = []
w_i_value = []


for w_i in w:
    predicted_data = np.dot(x_train,w_i) + b
    
    temp_cost = 0
    for i in range(len(x_train)):
        error_i = predicted_data[i] - y_train[i]
        error_sq = error_i**2
        cost_i = error_sq/len(x_train)

        temp_cost +=cost_i

    cost_total.append(temp_cost)
    w_i_value.append(w_i)

minimum_cost = min(cost_total)
minimum_cost_weight = w_i_value[np.argmin(cost_total)]


plt.plot(w_i_value,cost_total)
plt.xlabel('Weights')
plt.ylabel('Cost function value')
plt.title('Cost function vs weights')

plt.show()
```

```python
## Checking for bias for minimum cost wights

w = 210

b = np.array(range(0,200,10))


cost_total = []
b_i_value = []


for b_i in b:
    predicted_data = np.dot(x_train,w) + b_i
    
    temp_cost = 0
    for i in range(len(x_train)):
        error_i = predicted_data[i] - y_train[i]
        error_sq = error_i**2
        cost_i = error_sq/len(x_train)

        temp_cost +=cost_i

    cost_total.append(temp_cost)
    b_i_value.append(b_i)

minimum_cost = min(cost_total)
minimum_cost_bias = b_i_value[np.argmin(cost_total)]


plt.plot(b_i_value,cost_total)
plt.xlabel('Bias')
plt.ylabel('Cost function value')
plt.title('Cost function vs Bias')

plt.show()
```
